```python
#load dataset
import os
import numpy as np
import pandas as pd
from PIL import Image

captions_data = pd.read_csv('captions.txt')
captions_data = captions_data.dropna(subset=['caption'])
captions_data['caption'].fillna('', inplace=True)
image_filenames = captions_data['image']
captions = captions_data['caption']


# Path to the directory containing images
images_directory = 'Images/'


# Initialize lists to store data for the new CSV
new_image_names = []
new_captions = []
new_image_matrices = []
processed_image_names = set()  # Set to keep track of processed images

# Loop through each row in the existing data
for index, row in captions_data.iterrows():
    image_name = row['image']

    # Check if the image has already been processed
    if image_name not in processed_image_names:
      # Load and preprocess the image
      image_path = os.path.join(images_directory, image_name)
      image = Image.open(image_path)
      image = image.resize((224, 224))
      image_matrix = np.array(image)

      # Flatten the image matrix
      flattened_image_matrix = image_matrix.flatten()
        # Append data to the lists
      new_image_matrices.append(flattened_image_matrix)
    else:
      pass


      # Mark the image as processed
    processed_image_names.add(image_name)
    new_image_names.append(image_name)
    new_captions.append(caption)
    caption = row['caption']


# Create a new DataFrame
new_data = pd.DataFrame({
    'image': new_image_names,
    'caption': new_captions,
    'image_matrix': new_image_matrices
})

# Save the new DataFrame to a CSV file
new_csv_path = 'new_data.csv'
new_data.to_csv(new_csv_path, index=False)
```

```python
from PIL import Image
import time

def extract_image_features(filenames):
  image_features = []
  count = 0
  total_files = len(filenames)

  for filename in filenames:
    image = Image.open("Images/"+filename)
    image = image.resize((224, 224))

    image_array = np.array(image).flatten()

    image_features.append(image_array)

    count +=1
    print(f"Completed {count}/{total_files}")


  return np.array(image_features)

image_features = extract_image_features(captions_data['image'].unique())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(tokenizer=lambda x: x.split(), lowercase=False)
captions_matrix = vectorizer.fit_transform(captions)

```

```python
image_features
```

```python
X_train, X_val, y_train, y_val = train_test_split(image_features, captions_matrix, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report

svm_clf = SVC()
svm_clf.fit(X_train, y_train)
```

```python
predicated_captions = svm_clf.predict(X_val)

predicated_captions_text = vectorizer.inverse_transform(predicated_captions)
```
